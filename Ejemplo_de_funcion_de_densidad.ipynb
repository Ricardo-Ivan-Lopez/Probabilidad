```python
############## Ejemplo de funcion de densidad ##############
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros de la distribucin normal
mu = 0     # Media
sigma = 1  # Desviacin estndar

# Generar valores para x
x = np.linspace(-5, 5, 1000)

# Calcular la funcin de densidad
pdf = norm.pdf(x, mu, sigma)

# Graficar la funcin de densidad
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='PDF de la distribucin normal', color='blue')
plt.title('Funcin de Densidad de Probabilidad')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(mu, color='red', lw=0.8, ls='--', label='Media ($\mu$)')
plt.legend()
plt.grid()
plt.show()
```
