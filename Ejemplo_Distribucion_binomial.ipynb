```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

def coeficiente_binomial(n, k): #Calcula el coeficiente binomial (n sobre k)
    coeficiente = factorial(n) // (factorial(k) * factorial(n - k))
    return coeficiente

def probabilidad(k, n, p):  ##Calculamos la probabilidad de k éxitos en n experimentos con probabilidad p de éxito
    prob = coeficiente_binomial(n, k) * (p ** k) * ((1 - p) ** (n - k))
    return prob

def distribucion_binomial(n, p):
    # Definimos el rango de valores de k (número de éxitos)
    k = np.arange(0, n + 1)
    
    # Calculamos la distribución de probabilidad binomial
    prob = np.array([probabilidad(ki, n, p) for ki in k])
    
    print("Probabilidad de k de 0 hasta n = ")
    print(prob)
    
    # Graficar la distribución de probabilidad
    plt.figure(figsize=(10, 6))
    plt.stem(k, prob, linefmt='b-', markerfmt='bo', basefmt='r-')
    plt.title(f'Distribución Binomial - n={n}, p={p}')
    plt.xlabel('Número de Éxitos (k)')
    plt.ylabel('Probabilidad')
    plt.xticks(k)
    plt.grid(True)
    plt.show()

# Parámetros de la distribución binomial (usamos los datos del ejercicio de la clase)
n = 5  # Número total de experimentos
p = 1/6  # Probabilidad de éxito en cada experimento

# Llamar a la función para graficar la distribución
distribucion_binomial(n, p)
```
