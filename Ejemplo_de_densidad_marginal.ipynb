```python
############## Ejemplo de densidad marginal ##############

"""La función de densidad conjunta f_12(x, y) = 8xy está definida para 0 <= y <= x <= 1.
Fuera de este rango, la densidad conjunta es 0.
El objetivo es calcular las densidades marginales f_1(x) y f_2(y) a partir de f_12(x, y)."""

import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt

# Función para la densidad marginal f_1(x)
def f1(x):
    if 0 <= x <= 1:
        return integrate.quad(lambda y: 8 * x * y, 0, x)[0]  # Integrar sobre y de 0 a x
    else:
        return 0

# Función para la densidad marginal f_2(y)
def f2(y):
    if 0 <= y <= 1:
        return integrate.quad(lambda x: 8 * x * y, y, 1)[0]  # Integrar sobre x de y a 1
    else:
        return 0

# Generar valores para x y y
x_vals = np.linspace(0, 1, 100)
y_vals = np.linspace(0, 1, 100)

# Calcular f1(x) y f2(y)
f1_vals = [f1(x) for x in x_vals]
f2_vals = [f2(y) for y in y_vals]

# Graficar f1(x) y f2(y)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_vals, f1_vals, label='f1(x)', color='blue')
plt.title('Densidad marginal f1(x)')
plt.xlabel('x')
plt.ylabel('f1(x)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(y_vals, f2_vals, label='f2(y)', color='green')
plt.title('Densidad marginal f2(y)')
plt.xlabel('y')
plt.ylabel('f2(y)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
